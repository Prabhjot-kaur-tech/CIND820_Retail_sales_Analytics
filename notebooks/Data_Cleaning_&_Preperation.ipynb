## DATA CLEANING & PREPERATION


This section prepares the raw Online Retail dataset for reliable sales, product, country, trend, and customer behaviour analysis. The cleaning process removes duplicate rows, cancelled invoices, negative quantity records, and zero or negative unit price records. These steps are necessary because the project focuses on completed sales transactions.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
#upload raw excel data for analysis
df = pd.read_excel("/content/sample_data/Online_Retail_Raw_Data.xlsx")
df.head() # to check first five rows
```

```python
#create a copy of the raw data for cleaning
df_clean = df.copy()
```

```python
#remove duplicate rows
df_clean= df_clean.drop_duplicates()
```

```python
#remove cancelled invoices where InvoiceNo starts with "C"
df_clean = df_clean[~df_clean['InvoiceNo'].astype(str).str.startswith('C', na=False)]
```

```python
#Keep only positive  quantity values
df_clean = df_clean[df_clean['Quantity'] > 0]
```

```python
#keep only positive unit price values
df_clean = df_clean[df_clean['UnitPrice'] > 0]
```

```python
#Recalculate Revenue after cleaning
df_clean['Revenue']=df_clean['Quantity']*df_clean['UnitPrice']
```

```python
#Check cleaned dataset size
df_clean.shape
```

### Cleaning Summary

A cleaned version of the dataset was created by removing duplicate rows, cancelled invoices, negative quantity records, and zero or negative unit price records. These records were removed because they do not represent normal completed sales transactions and could distort revenue, product demand, and customer behaviour analysis.

The cleaned dataset will be used for revenue analysis, product analysis, country-level analysis, monthly trend analysis, and customer segmentation.

```python
#check cleaned missing values
df_clean.isnull().sum()
```

### Missing Values After Cleaning

After the initial cleaning process, missing values were reviewed again. Missing CustomerID values are especially important because customer-level analysis and RFM segmentation require valid customer identifiers. For general sales analysis, records without CustomerID may still be useful. However, for customer segmentation, records with missing CustomerID will be excluded.

```python
#create customer-level clean dataset for analysis
df_customer = df_clean.dropna(subset=['CustomerID']).copy()
df_customer['CustomerID'] = df_customer['CustomerID'].astype(int)
df_customer.shape
```

### Customer-Level Dataset

A separate customer-level dataset was created by removing records with missing CustomerID values. This dataset will be used for customer behaviour analysis and RFM segmentation. This separation is important because customer segmentation cannot be performed reliably without valid customer identifiers.

```python
#save clean dataset in csv format
df_clean.to_csv("online_retail_cleaned.csv", index=False)
df_customer.to_csv("online_retail_customer_cleaned.csv", index=False)
```
